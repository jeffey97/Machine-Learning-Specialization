```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x_train= np.array([1.0,2.0])
y_train= np.array([300.0,500.0])
print(f"x_train={x_train}") 
print(f"y_train={y_train}")
```

```python
m= x_train.shape[0]
print(m)
```

```python
i=0
x_i= x_train[i]
y_i=y_train[i]
print(f"x^{i}={x_i}", f"y^{i}={y_i}")
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title('Housing Prices')
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Price of the house (in 1000s dollars)')
plt.show()
```

```python
w=200
b=100
print(f'w:{w} b:{b}')
```

```python
def model_output(x, w, b ):
    m=len(x_train)
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=(w*x_train[i])+ b
    return f_wb

temp_f_wb= model_output(x_train, w, b)
print(temp_f_wb)
```

```python
plt.plot(x_train, temp_f_wb, c='b',label='Our Prediction')
plt.scatter(x_train, y_train, marker= 'x', c='r', label='Actual values')
plt.title('Housing Prices')
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Price of House (1000s dollars)')
plt.legend()
plt.show()
```
